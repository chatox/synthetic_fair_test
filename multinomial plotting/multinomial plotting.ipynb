```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

import pandas as pd
import numpy as np

mpl.rcParams.update({'font.size': 22, 'lines.linewidth': 3, 'lines.markersize': 15, 'font.family': 'Times New Roman'})
    # avoid type 3 (i.e. bitmap) fonts in figures
mpl.rcParams['ps.useafm'] = True
mpl.rcParams['pdf.use14corefonts'] = True
mpl.rcParams['text.usetex'] = True

data = pd.read_csv('FailprobIncreaseWithK_030303_01.csv')
x = data['x']
y = data['y']

data2 = pd.read_csv('FailprobIncreaseWithK_020503_01.csv')
x2 = data2['x']
y2 = data2['y']

data3 = pd.read_csv('FailprobIncreaseWithK_025025025025_01.csv')
x3 = data3['x']
y3 = data3['y']

data4 = pd.read_csv('FailprobIncreaseWithK_060202_01.csv')
x4 = data4['x']
y4 = data4['y']
_, ax = plt.subplots()
ax.plot(x, y, 'o', c='C1', markersize=5, markevery=3)
ax.plot(x2, y2, 'o', c='C2', markersize=5, markevery=3)
ax.plot(x3, y3, 'o', c='C3', markersize=5, markevery=3)
ax.plot(x4, y4, 'o', c='C4', markersize=5, markevery=3)
#ax.xaxis.set_major_locator(ticker.MultipleLocator(3))
plt.xlabel('k')
plt.ylabel('failprob')
plt.grid()
plt.legend(('p1=p2=p3=1/3', 'p1=0.2,p2=0.5,p3=0.2', 'p1=p2=p3=p4=0.25','p1=0.6,p2=p3=0.2'), loc="best", prop={'size': 10})
plt.savefig('FailprobIncreaseWithK_030303_01.png', bbox_inches='tight', dpi=300)
```
